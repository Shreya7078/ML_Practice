```python
import pandas as pd

# TV Shows
tv_shows = df[df['type'] == 'TV Show']
tv_genres = tv_shows['listed_in'].str.split(',').explode().str.strip()
tv_count = tv_genres.value_counts().reset_index()
tv_count.columns = ['Genre', 'TV Shows']

# Movies
movies = df[df['type'] == 'Movie']
movie_genres = movies['listed_in'].str.split(',').explode().str.strip()
movie_count = movie_genres.value_counts().reset_index()
print(movie_count)
movie_count.columns = ['Genre', 'Movies']
```

```python
comparison = pd.merge(tv_count, movie_count, on='Genre', how='outer').fillna(0)
comparison.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.barplot(x='TV Shows', y='Genre', data=comparison, color='skyblue', label='TV Shows')
sns.barplot(x='Movies', y='Genre', data=comparison, color='orange', label='Movies', alpha=0.7)

plt.xlabel('Count')
plt.ylabel('Genre')
plt.title('Popular Genres: TV Shows vs Movies')
plt.legend()
plt.show()
```

```python
#percentage of kids content

kids_genre=['Comedies','Children & Family Movies','Anime Series']
all_genres=df['listed_in'].str.split(',').explode().str.strip()
kids_content = all_genres[all_genres.isin(kids_genre)]
percentage = (len(kids_content) / len(all_genres)) * 100
print(f"Kids content percentage: {percentage:.2f}%")
```

```python
# 2010–2020 ke beech ka data
year_data = df[(df['release_year'] >= 2010) & (df['release_year'] <= 2020)]
yearly_count = year_data['release_year'].value_counts().sort_index()
print(yearly_count)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.lineplot(x=yearly_count.index, y=yearly_count.values, marker='o')
plt.title("Netflix Content Growth (2010–2020)")
plt.xlabel("Year")
plt.ylabel("Number of Shows / Movies")
plt.xticks(yearly_count.index)  # saal clearly dikhane ke liye
plt.show()
```
